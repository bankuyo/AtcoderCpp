## Takahashi and Meat

```c++
int num;
cin >> num;
vector<int> time(num);
for (int i = 0 ; i < num ; i++) cin >> time.at(i);

sort(time.begin(),time.end());
int sum1=0, sum2=time.at(num-1);

for (int i = 0 ; i < num-1 ; i++) sum1+=time.at(i);
for (int i = 0 ; i < num-1 ; i++){
    if (sum1 > sum2){
        sum1-=time.at(i);
        sum2+=time.at(i);
    }else break;
}
if (sum1 < sum2) cout << sum2 << endl;
else cout << sum1 << endl;
cout << sum1 << " " << sum2 << endl;
// for (int i = 0; i < num ; i++) cout << time.at(i) <<  " ";
// cout << endl;
```

```c++
int num;
cin >> num;
vector<int> time(num);
for (int i = 0 ; i < num ; i++) cin >> time.at(i);

int A;
for (int i = 0; i < num ; i++){
    A = A|1<<time.at(i);
}
vector<int> firstVec;
vector<int> secondVec;
for (int first = A ; ; first =(first-1) & A){
    for (int i = 0 ; i < num ; ++i){
        if (first & (1<<num)) firstVec.push_back(i);
    }
    for (int second = A ; ; second =(second-1) & A){
        for (int i = 0 ; i < num ; ++i){
            if (second & (1<<num)) secondVec.push_back(i);
        }
        cout << first << ": {";
        for (int i = 0; i < (int)firstVec.size(); ++i) {
            cout << firstVec[i] << " ";
        }
        cout << "}" << endl;
        if (!second) break;
    }
    if (!first) break;
}
```
