## Process
- Create the hash table for initial string : init
- Create the hash table for ideal string   : ideal

- Create the total hash table

- Loop through the strLength : i = 0 -> strLength - 1
    - Add  intiStr.at(strLength) to init
    - Add idealStr.at(strLength) to ideal
        - if  idealStr.at(strLength) doesn't exsit
            - set new
        - else 
            - if number of that character isn't enough
                - set new
            - else 
                - if the position is not changed
                    - set no
                - else
                    - set exsist
    - set total[totalChange][value]

```c++
// Data Input
int strLength, numLimit;
string initStr;
cin >> strLength >> numLimit >> initStr;

// Create the hash tables
map<char,  int>  initTable  ;
map<char,  int>  idealTable ;
map<int,string>  totalTable ;

// Create the ideal string
vector<char> idealStrVec;
string idealStr;
for (int i = 0 ; i < initStr.length() ; i++){
    idealStrVec.push_back(initStr.at(i));
}
sort(idealStrVec.begin(),idealStrVec.end());
for (int i = 0 ; i < initStr.length() ; i++){
    idealStr = idealStr+idealStrVec.at(i);
}

int count = 0;
int countChange = 0;
bool allChange = false;
vector<char> in;
vector<char> out;
string currentStr;
string leftStr = initStr.substr(0);

for (int i = 0 ; i < strLength ; i++){
    // Set the current Str
    currentStr = idealStr.substr(0,i+1);
    leftStr = initStr.substr(i+1);
    
    for (int j = 0 ; j < i+1 ; j++){
        initTable[initStr.at(j)] += 1;
        idealTable[idealStr.at(j)] += 1;
    }
    
    for (int j = 0 ; j < i+1 ; j++){
        if (idealTable[initStr.at(j)]< initTable[initStr.at(j)]){
            out.push_back(initStr.at(j));
            initTable[initStr.at(j)] = initTable[initStr.at(j)]-1;
        }
        if (idealTable[idealStr.at(j)]>initTable[idealStr.at(j)]){
            in.push_back(idealStr.at(j));
            idealTable[idealStr.at(j)] = idealTable[idealStr.at(j)]-1;
        }
    }
    sort(out.begin(),out.end());
    
    allChange = false;
    count = 0;
    for (int j = leftStr.length() -1 ; j >= 0; j--){
        for (int k = 0 ; k < in.size(); k++){
            if (leftStr.at(j) == in.at(k)){
                leftStr.at(j) = out.at(out.size()-1-count);
                in.erase(in.begin()+k);
                k--;
                count++;
            }
            if (count == out.size()){
                allChange = true;
                break;
            }
        }
        if (allChange) break;
    }
    
    currentStr = currentStr + leftStr;
    countChange = 0;
    for (int j = 0 ; j < currentStr.length(); j++){
        if (currentStr.at(j) != initStr.at(j)){
            countChange++;
        }
    }
    
    totalTable[countChange] = currentStr;
    initTable.clear();
    idealTable.clear();
    in.clear();
    out.clear();
}
totalTable[0] = initStr;
for (int i = 0 ; i < initStr.length();i++){
    if (totalTable[i].empty()){
        totalTable[i] = totalTable[i-1];
    }
}
totalTable[strLength] = idealStr;

// Debugging
for (int i = 0 ; i < initStr.length() ; i++){
    cout << idealStr.at(i) ;
}
cout << endl;
for (int i = 0 ; i <= initStr.length() ; i++){
    cout <<i << " : "<< totalTable[i] << endl;
}


// cout << endl;
cout << totalTable[numLimit] << endl;
```
