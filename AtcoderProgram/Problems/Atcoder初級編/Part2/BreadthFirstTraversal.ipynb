## Overview
- Problem : Find shortest time to arrive at goal from start
- Input :
    - raw : the number of raw
    - column : the number of column
    - startX, startY : The location (x,y) of starting point.  
    -  goalX,  goalY :The location (x,y) of ending point.  
    - grid : input of grid (raw * column)
        - "." : Which can pass
        - "#" : Which cannot pass
- Output : the shortest number to arrive at goal

## Example

```c++
   raw : 7  column : 8
startX : 2  startY : 2
 goalX : 4   goalY : 5
grid :
########  =>  ########
#......#  =>  #S.....#
#.######  =>  #.######
#..#...#  =>  #..#G..#
#..##..#  =>  #..##..#
##.....#  =>  ##.....#
########  =>  ########

Output => 11
```

## Consideration Part 1
- We might use breath first graph traversal
    - Need
        - Queue : To store the next candidate location
        - Map   : To cache the alredy visited spots
            - key : Node
            - value : current shortest path
        - Option : To extract shortest path
            - Map : { key = Node : value = previous Node }
                - default previous Node is null
   
## Consideration Part 2
- Create queue : candidate
    - Set to the startX and startY pair
- Create Map : shortestTable
    - Insert {start location : 0 }
- Create Map : adjacentTable ( **Option** )
    - Insert {start location : start location}
- while candidate is not empty
    - Take current Node from candidate
    - Take current shortest value from shortestTable
    - Find the next candidate
        - if next candidate is already visited
            - Pass
        - Otherwise
            - if next Node is the goal
                - Set the shortest path to be current shortest +1
                - Set the adjacentTable[goal] to be current Node (**Option**)
                - break
            - Add to queue
            - Add to adjacentMap
                - key : next Node
                - value : current Node
            - Set shortestTable[next Node] to be current shortest + 1

- Traversal end
- Return shortest value
## Option - Bringing back the path 
- Create the path stack or vector : stack
- Set current Node to be goal Node
- While it is not start Point
    - Push back to stack
    - Set current Node to be previous Node (adjacentTablr)
- While stack is not empty
    - Print the top of stack
    - pop from stack

```c++
// Data Input
int    raw , column  ;
vector<int> start(2) ;
vector<int>  goal(2) ; 
cin >> raw >> column;
// Find the bug 
cin >> start.at(1) >> start.at(0) ; // cin >> start.at(0) >> start.at(1) ;
cin >> goal.at(1)  >> goal.at(0)  ; // cin >> goal.at(0)  >> goal.at(1)  ;

start.at(0) = start.at(0)-1;
start.at(1) = start.at(1)-1;
goal.at(0) = goal.at(0)-1;
goal.at(1) = goal.at(1)-1;

vector<string> grid(raw);
for (int i = 0 ; i < raw ; i++) cin >> grid.at(i);

// Create queue : candidate
queue<vector<int>> candidate;
candidate.push(start);               // (1, 1)

//Create Map : shortestTable
map<vector<int>, int> shortestTable;
shortestTable[start] = 0;           // (1, 1) : 0

// Create Map : adjacentTable
map<vector<int>, vector<int>> adjacentTable;
adjacentTable[start] = start;       // (1, 1) : (1, 1)

// Declare currentNode and nextNode
vector<int> currentNode, nextNode;

// Declare two vector to find next node
vector<int> aroundX = {-1, 0, 1, 0};
vector<int> aroundY = {0, 1, 0, -1};

// while candidate is not empty
while (!candidate.empty()){
    // Take current Node from candidate
    currentNode = candidate.front();          // (1, 1) => (1, 2)
    candidate.pop();                          // 
    
    // Take current shortest value from shortestTable
    // Find the next candidate
    for (int i = 0 ; i < 4 ; i++){
        if (currentNode.at(0) + aroundX.at(i) >= 0 && currentNode.at(0) + aroundX.at(i) < column &&
            currentNode.at(1) + aroundY.at(i) >= 0 && currentNode.at(1) + aroundY.at(i) < raw       ){
            if (grid.at(currentNode.at(1) + aroundY.at(i)).at(currentNode.at(0) + aroundX.at(i)) == '.'){  // (2, 1)
                nextNode = { currentNode.at(0) + aroundX.at(i), currentNode.at(1)+aroundY.at(i)};         // (1, 2)            
                 // if next Node is not visited
                if (shortestTable[nextNode]==0 && nextNode != start){
                    // Debug2
                    // cout << grid.at(currentNode.at(1) + aroundY.at(i)).at(currentNode.at(0) + aroundX.at(i)) <<" ";
                    // cout << nextNode.at(0) << " "<< nextNode.at(1)<< endl; 
                    
                    // Add to queue
                    candidate.push(nextNode);
                    // Add to adjacentMap
                    adjacentTable[nextNode] = currentNode;
                    // Set shortestTable[next Node] to be current shortest + 1
                    shortestTable[nextNode] = shortestTable[currentNode] + 1;
                    // if next Node is the goal
                    if (nextNode == goal){ 
                        // Debug 1
                        // cout << "shortestTable[goal]" << endl; 
                        break;
                    }
                }
            }
        }
    }
}

// Debug 3
// for (int i = 0 ; i < raw ; i++) cout << grid.at(i) <<endl;
// Return shortest value
cout << shortestTable[goal] << endl;
```

# Option Extract Path

## Pseudocode
- Create the path stack or vector : stack
- Set current Node to be goal Node
- While it is not start Point
    - Push back to stack
    - Set current Node to be previous Node (adjacentTablr)
- While stack is not empty
    - Print the top of stack
    - pop from stack

```c++
// Data Input
int    raw , column  ;
vector<int> start(2) ;
vector<int>  goal(2) ; 
cin >> raw >> column;
cin >> start.at(1) >> start.at(0) ;
cin >> goal.at(1)  >> goal.at(0)  ;
start.at(0) = start.at(0)-1;
start.at(1) = start.at(1)-1;
goal.at(0) = goal.at(0)-1;
goal.at(1) = goal.at(1)-1;
vector<string> grid(raw);
for (int i = 0 ; i < raw ; i++) cin >> grid.at(i);


// Main Part
queue<vector<int>> candidate;
map<vector<int>, int> shortestTable;
map<vector<int>, vector<int>> adjacentTable;
vector<int> currentNode, nextNode;

vector<int> aroundX = {-1, 0, 1, 0};
vector<int> aroundY = {0, 1, 0, -1};

// Initialize
candidate.push(start);
shortestTable[start] = 0;
adjacentTable[start] = start; 

while (!candidate.empty()){
    currentNode = candidate.front();          
    candidate.pop();             
    
    for (int i = 0 ; i < 4 ; i++){
        if (currentNode.at(0) + aroundX.at(i) >= 0 && currentNode.at(0) + aroundX.at(i) < column &&
            currentNode.at(1) + aroundY.at(i) >= 0 && currentNode.at(1) + aroundY.at(i) < raw       ){
            
            if (grid.at(currentNode.at(1) + aroundY.at(i)).at(currentNode.at(0) + aroundX.at(i)) == '.'){
                
                nextNode = { currentNode.at(0) + aroundX.at(i), currentNode.at(1)+aroundY.at(i)}; 
                
                if (shortestTable[nextNode]==0 && nextNode != start){
                    candidate.push(nextNode);
                    adjacentTable[nextNode] = currentNode;
                    shortestTable[nextNode] = shortestTable[currentNode] + 1;
                    if (nextNode == goal) break;
                }
            }
        }
    }
}

// Optional Extract Path from adjacentTable
// Create the path stack or vector : stack
stack<vector<int>> stack;

// Set current Node to be goal Node
currentNode = goal;

// While it is not start Point
while (currentNode != start){
    // Push back to stack
    stack.push(currentNode);
    // Set current Node to be previous Node (adjacentTablr)
    currentNode = adjacentTable[currentNode];
}

// Add the start at top
stack.push(start);

int count = 1;
// While stack is not empty
while (!stack.empty()){
    // Print the top of stack
    cout << "Node "<< count <<" : " << stack.top().at(0) << " " << stack.top().at(1)<< endl ;
    // pop from stack
    stack.pop();
    count++;
}
```
